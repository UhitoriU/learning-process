```python
import os
from openai import OpenAI
```

```python
def openai_embedding(text:str,model:str=None):
    client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"),base_url=os.getenv("BASE_URL"))

    if model == None:
        model = "text-embedding-3-small"

    response = client.embeddings.create(
        input=text,
        model=model
    )
    return response
```

```python
response = openai_embedding(text='要生成embedding的文本，字符串形式')
print(response)
```

```python
#通过response的object来获取embedding的类型
print(f'返回的embedding类型为:{response.object}')
```

```python
#embedding存放在data中
print(f'embedding长度为:{len(response.data[0].embedding)}')
print(f'embedding(前10)为:{response.data[0].embedding[:10]}')
```

```python
#查看此次embedding的模型及token使用情况
print(f'本次embedding model为:{response.model}')
print(f'本次token使用情况为:{response.usage}')
```
